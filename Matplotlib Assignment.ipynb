Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

===>
Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations. It provides a wide range of tools for producing high-quality plots, charts, and figures, making it one of the most widely used data visualization libraries in the Python ecosystem

Why is Matplotlib used?
Matplotlib is used for various data visualization tasks, including but not limited to:

Exploring and understanding data: Matplotlib allows users to quickly visualize data to gain insights and identify patterns.

Communicating results: Matplotlib enables the creation of publication-quality visualizations for scientific papers, reports, presentations, and more.

Data-driven storytelling: Matplotlib helps to tell compelling stories using data visualizations, making complex information more accessible to the audience.

Supporting data analysis: Matplotlib is often used in conjunction with other data analysis libraries (e.g., Pandas, Numpy) to visualize and analyze data efficiently.

Five plots that can be plotted using the Pyplot module of Matplotlib:


1)Line plot: Used to display data as points (markers) connected by straight lines, commonly used to show trends and relationships.
    
2)Scatter plot: Used to visualize the relationship between two numerical variables, displaying individual data points as dots.
    
3_Bar plot: Used to display categorical data with rectangular bars, commonly used to compare different categories or groups.
    
4)Histogram: Used to represent the distribution of a continuous variable by dividing the data into bins and counting the occurrences in each bin.
    
5)Pie chart: Used to represent the composition of a whole by dividing it into slices, where each slice represents a proportion of the whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

==>A scatter plot is a set of dotted points representing individual data pieces on the horizontal and vertical axis. In a graph in which the values of two variables are plotted along the X-axis and Y-axis, the pattern of the resulting points reveals a correlation between them.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Data")
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

==>The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange multiple plots in a grid-like structure, specifying the number of rows, number of columns, and the position of each subplot within the grid. The subplot() function is helpful when you want to display multiple related plots side by side for easy comparison or visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Python")
plt.grid()

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Statistics")
plt.grid()


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Machine Learning")
plt.grid()


x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,3,4)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Deep Learning")
plt.grid()

plt.suptitle("DATA SCIENTIST")
plt.plot(x,y)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

==>Bar plot (also known as bar chart) is a data visualization technique used to compare different categories or groups by representing them as rectangular bars. 


Why is a bar plot used?
Bar plots are used for the following purposes:

1)Comparing values: Bar plots make it easy to visually compare the magnitudes of different categories or groups.
    
2)Showing distributions: Bar plots can display the frequency or occurrence of categorical data, such as the number of occurrences of each category.
    
3)Identifying trends: Bar plots can reveal patterns and trends within categorical data, helping to make data-driven decisions.
    
4)Presenting survey results: Bar plots are often used to present survey results, showing responses to different options.


```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("Company Names")
plt.ylabel("Valuation")

plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

==>
A box plot (also known as box-and-whisker plot) is a graphical representation of the distribution of a dataset, showing the median, quartiles, and potential outliers. It displays the data's summary statistics in a compact and easy-to-understand format. A box plot consists of a rectangular "box" that represents the interquartile range (IQR), with a line inside the box indicating the median value. "Whiskers" extend from the box to show the range of the data, excluding potential outliers. Outliers are individual data points that are significantly different from the rest of the data.

Why is a box plot used?

Box plots are used for the following purposes:

1)Visualizing data distribution: Box plots provide a clear picture of the spread and central tendency of the data.
    
2)Identifying outliers: Box plots help in identifying extreme values or potential outliers in the dataset.
    
3)Comparing multiple datasets: Box plots are useful for comparing the distribution of different datasets side by side.
    
4)Highlighting data variability: Box plots show the variability and skewness of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("BOX PLOT")

plt.boxplot(data)
plt.show()
```
